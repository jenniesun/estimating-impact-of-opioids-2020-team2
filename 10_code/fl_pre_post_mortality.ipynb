```python
import pandas as pd
```

### Florida Pre-Policy Mortality

```python
# load mortality the data

death03 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2003.txt",
    error_bad_lines=False,
    sep="\t",
)
death04 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2004.txt",
    error_bad_lines=False,
    sep="\t",
)
death05 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2005.txt",
    error_bad_lines=False,
    sep="\t",
)
death06 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2006.txt",
    error_bad_lines=False,
    sep="\t",
)
death07 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2007.txt",
    error_bad_lines=False,
    sep="\t",
)
death08 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2008.txt",
    error_bad_lines=False,
    sep="\t",
)
death09 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2009.txt",
    error_bad_lines=False,
    sep="\t",
)
death10 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2010.txt",
    error_bad_lines=False,
    sep="\t",
)
death11 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2011.txt",
    error_bad_lines=False,
    sep="\t",
)
death12 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2012.txt",
    error_bad_lines=False,
    sep="\t",
)
death13 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2013.txt",
    error_bad_lines=False,
    sep="\t",
)
death14 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2014.txt",
    error_bad_lines=False,
    sep="\t",
)
death15 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2015.txt",
    error_bad_lines=False,
    sep="\t",
)
```

```python
# concatinate all pre-policy data

pre_policy = pd.concat(
    [death03, death04, death05, death06, death07, death08, death09], ignore_index=True
)
```

```python
pre_policy
```

```python
# Subset to Florida data

pre_policy[["County1", "State"]] = pre_policy.County.str.split(",", expand=True)
pre_policy_fl = pre_policy[pre_policy["State"] == " FL"]
pre_policy_fl.head()
```

```python
# drop columnes: Notes, County, County Code, Year Code; rename County1 to County

pre_policy_fl = pre_policy_fl.drop(
    ["Notes", "County", "County Code", "Year Code"], axis=1
)
pre_policy_fl = pre_policy_fl.rename(columns={"County1": "County"})
```

```python
pre_policy_fl.head()
```

```python
len(pre_policy_fl["County"].unique())
```

```python
# obtain codes for mortality for drug related causes

pre_policy_fl.groupby(
    ["Drug/Alcohol Induced Cause Code", "Drug/Alcohol Induced Cause"], as_index=False
).mean()
```

```python
# subset to mortality to drug related causes

pre_policy_fl_drug = pre_policy_fl[
    (pre_policy_fl["Drug/Alcohol Induced Cause Code"] == "D1")
    | (pre_policy_fl["Drug/Alcohol Induced Cause Code"] == "D2")
    | (pre_policy_fl["Drug/Alcohol Induced Cause Code"] == "D4")
    | (pre_policy_fl["Drug/Alcohol Induced Cause Code"] == "D9")
]
```

```python
pre_policy_fl_drug.head()
```

```python
# arrange by unique identifier: Year-State-County

pre_policy_fl_drug = pre_policy_fl_drug.groupby(
    ["Year", "State", "County"], as_index=False
).sum()
pre_policy_fl_drug.head()
```

```python
# convert year datatype from float to int

pre_policy_fl_drug["Year"] = pre_policy_fl_drug["Year"].astype(int)
pre_policy_fl_drug.head()
```

### Florida Post-Policy Mortality

```python
# concatinate all post-policy data

post_policy = pd.concat(
    [death10, death11, death12, death13, death14, death15], ignore_index=True
)
```

```python
post_policy
```

```python
# subset to Florida data

post_policy[["County1", "State"]] = post_policy.County.str.split(",", expand=True)
post_policy_fl = post_policy[post_policy["State"] == " FL"]
post_policy_fl.head()
```

```python
# drop columnes: Notes, County, County Code, Year Code; rename County1 to County

post_policy_fl = post_policy_fl.drop(
    ["Notes", "County", "County Code", "Year Code"], axis=1
)
post_policy_fl = post_policy_fl.rename(columns={"County1": "County"})
post_policy_fl.head()
```

```python
# subset to mortality to drug related causes

post_policy_fl_drug = post_policy_fl[
    (post_policy_fl["Drug/Alcohol Induced Cause Code"] == "D1")
    | (post_policy_fl["Drug/Alcohol Induced Cause Code"] == "D2")
    | (post_policy_fl["Drug/Alcohol Induced Cause Code"] == "D4")
    | (post_policy_fl["Drug/Alcohol Induced Cause Code"] == "D9")
]
post_policy_fl_drug.head()
```

```python
import warnings

warnings.filterwarnings("ignore")

# convert column Deaths to float datatype to prepare for sum in the next step

post_policy_fl_drug["Deaths"] = post_policy_fl_drug["Deaths"].astype(float)
```

```python
# arrange by unique identifier: Year-State-County

post_policy_fl_drug = post_policy_fl_drug.groupby(
    ["Year", "State", "County"], as_index=False
).sum()

post_policy_fl_drug.head()
```

```python
# convert year datatype from float to int

post_policy_fl_drug["Year"] = post_policy_fl_drug["Year"].astype(int)
post_policy_fl_drug.head()
```

### Merge Mortality Data with Population Data to Calculate Mortality Rate

```python
# load pre-policy population the data

fl_pop_pre_policy = pd.read_excel("./Population/fl_pop_2000-2010.xls", header=[3])
```

```python
fl_pop_pre_policy
```

```python
# keep columns with population data from 2003 to 2009; rename 'Unnamed: 0' to County

fl_pop_pre_policy = fl_pop_pre_policy.drop(
    ["Unnamed: 1", 2000, 2001, 2002, "Unnamed: 12", "Unnamed: 13"], axis=1
)
fl_pop_pre_policy = fl_pop_pre_policy.rename(columns={"Unnamed: 0": "County"})
fl_pop_pre_policy
```

```python
# check last n rows with NaN values to determine the number of rows to drop

fl_pop_pre_policy.iloc[-9:]
```

```python
# remove first row - total state population and last 8 rows with NaN values

fl_pop_pre_policy.drop(fl_pop_pre_policy.head(1).index, inplace=True)
fl_pop_pre_policy.drop(fl_pop_pre_policy.tail(8).index, inplace=True)
fl_pop_pre_policy
```

```python
# remove the first character '.' in County Names

fl_pop_pre_policy["County"] = fl_pop_pre_policy["County"].str[1:]
fl_pop_pre_policy
```

```python
# melt the data and arrange columns by unique identifier: Year-County

melted = fl_pop_pre_policy.melt(id_vars=["County"])
melted = melted.rename(columns={"variable": "Year", "value": "Population"})
melted = melted.groupby(["Year", "County"], as_index=False).sum()
melted
```

```python
# merge population data with pre policy drug mortality data

merged = pre_policy_fl_drug.merge(melted, how="left")
merged
```

```python
# create county mortality rate column

merged["Mortality Rate"] = merged["Deaths"] / merged["Population"]
merged
```
