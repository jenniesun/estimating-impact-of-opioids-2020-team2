```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ARdata = pd.read_csv("/Users/dapoadegbile/estimating-impact-of-opioids-2020-team2/20_intermediate_files/AR_shipment_and_pop.csv", index_col = 0)
NMdata = pd.read_csv("/Users/dapoadegbile/estimating-impact-of-opioids-2020-team2/20_intermediate_files/NM_shipment_and_pop.csv", index_col = 0)
PAdata = pd.read_csv("/Users/dapoadegbile/estimating-impact-of-opioids-2020-team2/20_intermediate_files/PA_shipment_and_pop.csv", index_col = 0)
```

```python
PAdata.head(5)
```

```python
newARdata = ARdata.groupby(['Year', 'BUYER_STATE', "County", "Population"])["MORPHINE_EQUIV_IN_MG"].sum().reset_index()
newARdata.head(5)
```

```python
newNMdata = NMdata.groupby(['Year', 'BUYER_STATE', "County", "Population"])["MORPHINE_EQUIV_IN_MG"].sum().reset_index()
newNMdata.head(5)
```

```python
newPAdata = PAdata.groupby(['Year', 'BUYER_STATE', "County", "Population"])["MORPHINE_EQUIV_IN_MG"].sum().reset_index()
PAdata.head(5)
```

```python
newPAdata[["Population"]] = newPAdata["Population"].str.replace(',', '').astype(float)
newARdata[["Population"]] = newARdata["Population"].str.replace(',', '').astype(float)
newNMdata[["Population"]] = newNMdata["Population"].str.replace(',', '').astype(float)

```

```python
comparisonStates = pd.concat([newARdata, newNMdata, newPAdata])
comparisonStates.sample(10)
```

```python
comparisonStates["MORPH_POP_RATIO"] = comparisonStates["MORPHINE_EQUIV_IN_MG"] / comparisonStates["Population"] 
```

```python
comparisonStates.sample(10)
```
